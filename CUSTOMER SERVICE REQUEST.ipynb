# Elapsed Time to solve created compliant

```python
df1["ElapsedTime"]= df1["Closed Date"].values-df1["Created Date"].values
```

```python
df1["ElapsedTime"]
```

```python
ElapsedTime=[]
for x in (df1["Closed Date"]-df1["Created Date"]):
    close=x.total_seconds()/60/60
    ElapsedTime.append(close)
df1["ElapsedTime"]=ElapsedTime
```

# Kruskal Wallis H Test

## Null Hypothesis: The average response time across types of complaints are not different
## Alternate Hypothesis: The average response time across types of complaints are different

```python
from scipy import stats
```

```python
dataset={}
for i in df2["Complaint Type"].unique():
    dataset[i]=df2[df2["Complaint Type"]==i]["ElapsedTime"]
```

```python
dataset.keys()
```

```python
stats, p=stats.kruskal(dataset["Blocked Driveway"], dataset["Illegal Parking"], dataset["Noise - Street/Sidewalk"], dataset["Noise - Commercial"], dataset["Derelict Vehicle"], dataset["Noise - Vehicle"])

alpha=0.05
if p>0.05:
    print("Null Hypothesis is Accepted")
else:
    print("Null Hypothesis is Rejected")
```

```python
stats, p
```

## Null Hypothesis: The average response time across types of locations are not different
## Alternate Hypothesis: The average response time across types of Locations are different

```python
dataset={}
for i in df2["Location Type"].unique():
    dataset[i]=df2[df2["Location Type"]==i]["ElapsedTime"]
```

```python
dataset.keys()
```

```python
stats, p=stats.kruskal(dataset["Street/Sidewalk"], dataset["Club/Bar/Restaurant"], dataset["Store/Commercial"], dataset["Residential Building/House"], dataset["Park/Playground"], dataset["House of Worship"])

alpha=0.05
if p>0.05:
    print("Null Hypothesis is Accepted")
else:
    print("Null Hypothesis is Rejected")
```

```python
stats, p
```

## Null Hypothesis: The average response time across cities are not different
## Alternate Hypothesis: The average response time across cities are different

```python
dataset={}
for i in df2["City"].unique():
    dataset[i]=df2[df2["City"]==i]["ElapsedTime"]
```

```python
dataset.keys()
```

```python
stats, p=stats.kruskal(dataset["BROOKLYN"], dataset["NEW YORK"], dataset["BRONX"], dataset["STATEN ISLAND"], dataset["JAMAICA"], dataset["ASTORIA"])

alpha=0.05
if p>0.05:
    print("Null Hypothesis is Accepted")
else:
    print("Null Hypothesis is Rejected")
```

```python
stats, p
```
